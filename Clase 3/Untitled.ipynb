```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Automobile.csv")

print(df.head())
```

Grafico de Dispersión: MPG VS WEIGHT

```python

plt.figure(figsize=(8,5))
plt.scatter(df["weight"], df["mpg"], color="steelblue", alpha=0.7)
plt.xlabel("Peso del automóvil")
plt.ylabel("Millas por galón (mpg)")
plt.title("Relación entre peso y consumo de combustible")
plt.grid(True)
plt.show()
```

Histograma del consumo de Combustible

```python

plt.figure(figsize=(8,5))
plt.hist(df["mpg"], bins=20, color="green", edgecolor="black")
plt.xlabel("Millas por galón (mpg)")
plt.ylabel("Frecuencia")
plt.title("Distribución del consumo de combustible")
plt.grid(True)
plt.show()
```

Boxplot de Cylinders y origin

```python

df.boxplot(column='cylinders', by='origin', grid=True)

plt.title('Distribución de los cilindros por origen')
plt.suptitle('')
plt.xlabel('Origen del automóvil')
plt.ylabel('Número de cilindros')

plt.show()
```

Gráfico de barras: model_year vs origin

```python

conteo = df.groupby(["model_year", "origin"]).size().unstack()

conteo.plot(kind="bar", figsize=(10,6))
plt.xlabel("Año del modelo")
plt.ylabel("Cantidad de automóviles")
plt.title("Distribución de automóviles por año y origen")
plt.legend(title="Origen")
plt.grid(axis="y")
plt.show()
```
