```python
import numpy as np

matriz1 = np.random.randint(1, 11, size=(3, 3))
matriz2 = np.random.randint(1, 11, size=(3, 3))

producto = np.dot(matriz1, matriz2)

print("Matriz 1:\n", matriz1)
print("Matriz 2:\n", matriz2)
print("Producto:\n", producto)
```

```python
import numpy as np
array_random = np.random.randint(1, 101, 15)

maximo = array_random.max()
minimo = array_random.min()
promedio = array_random.mean()

print("Array:", array_random)
print("Máximo:", maximo)
print("Mínimo:", minimo)
print("Promedio:", promedio)
```

```python
import numpy as np
matriz_5x5 = np.arange(1, 26).reshape(5, 5)
diagonal = np.diag(matriz_5x5)

print("Matriz:\n", matriz_5x5)
print("Diagonal:", diagonal)
```

```python
import numpy as np
matriz = np.random.randint(1, 21, size=(5, 5))

suma_filas = matriz.sum(axis=1)
suma_columnas = matriz.sum(axis=0)

print("Matriz:\n", matriz)
print("Suma por filas:", suma_filas)
print("Suma por columnas:", suma_columnas)
```
